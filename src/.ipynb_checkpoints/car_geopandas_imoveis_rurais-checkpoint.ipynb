```python
## Importação dos pacotes geopandas e matplotlib
import matplotlib.pyplot as plt
import geopandas
import seaborn as sns

## Leitura do arquivo com as informações dos imóveis rurais
path = '../data/imoveis_rurais.geojson'
df = geopandas.read_file(path)
```

```python
## Informações sobre o conteúdo do arquivo
df.info()
```

```python
## Imagem dos dados em uma figura (plotagem)
df.plot(figsize=(30,15))
```

```python
## Distribuição dos imóveis pequenos
sns.boxplot(x=df['NUM_AREA'])
plt.show()
```

```python
## Distribuição dos imóveis pequenos
pequenos = df[df['NUM_AREA'] <= 40]
sns.boxplot(x=pequenos['NUM_AREA'])
plt.show()
```

```python
## Distribuição dos imóveis pequenos
pequenos = df[df['NUM_AREA'] <= 40]
#medios["NUM_AREA"].round(0).value_counts()

sns.swarmplot(x = pequenos["NUM_AREA"].round(0))  
##sns.(x=pequenos['NUM_AREA'])
#pequenos
#sns.displot(
#    pequenos, x="NUM_AREA", col="NUM_AREA",
#    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
#)
#plt.show()
```

```python
## Distribuição dos imóveis médios (> 40ha e < 150)
medios = df[(df['NUM_AREA'] > 40) & (df['NUM_AREA'] <= 150)]
sns.boxplot(x=medios['NUM_AREA'])
plt.show()
```

```python
#sns.swarmplot(x = medios["NUM_AREA"]) 
medios["NUM_AREA"].round(0).value_counts()
```

```python
## Distribuição dos grandes imóveis
grandes = df[df['NUM_AREA'] > 150]
sns.boxplot(x=grandes['NUM_AREA'])
plt.show()
```

```python
#sns.swarmplot(x = grandes["NUM_AREA"]) 
grandes["NUM_AREA"].round(1).value_counts()
```

```python
df.plot(column='NUM_AREA', cmap='BuPu', edgecolor='k', legend=True, figsize=(30,15),legend_kwds={'label': "Área(Ha)",
    'orientation': "vertical"})
plt.savefig('capivari.png')
```

```python
pequenos=0
medios=0
grandes=0

for a in df['NUM_AREA']:
    if a <= 40:
        pequenos+=1
    if (a > 40) & (a <= 150):
        medios+=1
    if (a > 150): 
        grandes+=1

print("Quantidade de imóveis pequenos (<=40ha):", pequenos)
print("Quantidade de imóveis médios (>40ha e <=150ha):", medios)
print("Quantidade de imóveis grandes (>150ha):", grandes)
```
